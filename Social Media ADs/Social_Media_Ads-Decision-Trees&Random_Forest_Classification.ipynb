```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline
```

```
dataset = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Machine Learning A-Z/Part 3 - Classification/Section 18 - Naive Bayes/Naive_Bayes/Social_Network_Ads.csv")
```

```
X = dataset.iloc[:,[2,3]]
#X = X.values.reshape(-1,1)
y = dataset.iloc[:,4]
#y = y.values.reshape(-1,1)
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=0)
```

```
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

# Decision Tree Classifier

```
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'entropy')
model.fit(X_train,y_train)
```

```
y_pred = model.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```
cm
```

### There is 61 & 30 = 91 is Correct Predictions 7 & 2 = 9 Incorrect Predictions

# Random Forest 

```
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10,criterion = 'entropy',random_state = 0)
model.fit(X_train,y_train)
```

```
y_pred2 = model.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred2)
```

```
cm
```

### There is 63 & 27 = 90 is Correct Predictions 5 & 5 = 10 Incorrect Predictions
